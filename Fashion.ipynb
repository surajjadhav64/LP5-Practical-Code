````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
````

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization
from sklearn.metrics import classification_report
````

````python
train_df = pd.read_csv('fashion-mnist_train.csv')
test_df = pd.read_csv('fashion-mnist_test.csv')
````

````python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
````

````python
x_train = train_df.iloc[:, 1:].values.reshape(-1, 28, 28, 1) / 255.0
y_train = train_df.iloc[:, 0].values
x_test = test_df.iloc[:, 1:].values.reshape(-1, 28, 28, 1) / 255.0
y_test = test_df.iloc[:, 0].values
````

````python
plt.figure(figsize=(8, 8))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([]), plt.yticks([]), plt.grid(False)
    plt.imshow(x_train[i], cmap='binary')
    plt.xlabel(class_names[y_train[i]])
plt.tight_layout()
plt.show()
````

````python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(28, 28, 1)),
    BatchNormalization(),
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2, 2)), Dropout(0.25),

    Conv2D(128, (3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D((2, 2)), Dropout(0.25),

    Flatten(),
    Dense(128, activation='relu'), Dropout(0.5),
    Dense(10, activation='softmax')
])

# Compile and train the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()
````

````python
model.fit(x_train, y_train, epochs=50, batch_size=5000, validation_split=0.05)
````

````python
model.evaluate(x_test, y_test)
y_pred = model.predict(x_test).argmax(axis=-1)
````

````python
plt.figure(figsize=(8, 8))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks([]), plt.yticks([]), plt.grid(False)
    plt.imshow(x_test[i], cmap='binary')
    plt.title(f"Pred: {class_names[y_pred[i]]}")
plt.tight_layout()
plt.show()
````

````python
print(classification_report(y_test, y_pred, target_names=class_names))
````

````python
# Fashion MNIST CNN Project – Code with Line-by-Line Comments

# Import NumPy library for numerical operations and array handling
import numpy as np

# Import Pandas library for reading and handling datasets
import pandas as pd

# Import Seaborn library for data visualization
import seaborn as sns

# Import Matplotlib library for plotting graphs and images
import matplotlib.pyplot as plt

# Import warnings module to manage warning messages
import warnings

# Ignore all warning messages during execution
warnings.filterwarnings('ignore')
```

```python
# Import Sequential model from Keras to build neural network layer-by-layer
from tensorflow.keras.models import Sequential

# Import CNN layers used for image classification
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization

# Import classification report for evaluating model performance
from sklearn.metrics import classification_report
```

```python
# Read training dataset CSV file into DataFrame
train_df = pd.read_csv('fashion-mnist_train.csv')

# Read testing dataset CSV file into DataFrame
test_df = pd.read_csv('fashion-mnist_test.csv')
```

```python
# Store class labels of Fashion MNIST dataset in a list
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
# Extract pixel values from training data, reshape into 28x28 images, and normalize values between 0 and 1
x_train = train_df.iloc[:, 1:].values.reshape(-1, 28, 28, 1) / 255.0

# Extract training labels from first column
y_train = train_df.iloc[:, 0].values

# Extract pixel values from testing data, reshape into 28x28 images, and normalize values
x_test = test_df.iloc[:, 1:].values.reshape(-1, 28, 28, 1) / 255.0

# Extract testing labels from first column
y_test = test_df.iloc[:, 0].values
```

```python
# Create a figure with size 8x8 inches
plt.figure(figsize=(8, 8))

# Loop through first 25 images
for i in range(25):
    
    # Create subplot in 5x5 grid
    plt.subplot(5, 5, i+1)
    
    # Remove x-axis and y-axis ticks
    plt.xticks([]), plt.yticks([])
    
    # Disable grid display
    plt.grid(False)
    
    # Display image in binary color map
    plt.imshow(x_train[i], cmap='binary')
    
    # Display corresponding class label below image
    plt.xlabel(class_names[y_train[i]])

# Show all plotted images
plt.show()
```

```python
# Create Sequential CNN model
model = Sequential([
    
    # First convolutional layer with 32 filters of size 3x3 and ReLU activation
    Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=(28, 28, 1)),
    
    # Normalize activations to improve training speed and stability
    BatchNormalization(),
    
    # Second convolutional layer with 64 filters
    Conv2D(64, (3, 3), activation='relu', padding='same'),
    
    # Apply batch normalization again
    BatchNormalization(),
    
    # Reduce image dimensions using max pooling
    MaxPooling2D((2, 2)),
    
    # Randomly drop neurons to reduce overfitting
    Dropout(0.25),
    
    # Third convolutional layer with 128 filters
    Conv2D(128, (3, 3), activation='relu', padding='same'),
    
    # Normalize activations
    BatchNormalization(),
    
    # Reduce dimensions again using max pooling
    MaxPooling2D((2, 2)),
    
    # Apply dropout for regularization
    Dropout(0.25),
    
    # Convert multidimensional feature maps into 1D vector
    Flatten(),
    
    # Fully connected dense layer with 128 neurons
    Dense(128, activation='relu'),
    
    # Apply dropout to avoid overfitting
    Dropout(0.5),
    
    # Output layer with 10 neurons for 10 classes using softmax activation
    Dense(10, activation='softmax')
])
```

```python
# Compile model by defining optimizer, loss function, and evaluation metric
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the CNN model on training dataset
model.fit(x_train,
          y_train,
          epochs=50,
          batch_size=5000,
          validation_split=0.05)
```

```python
# Evaluate model performance on testing dataset
model.evaluate(x_test, y_test)

# Predict output classes for test images
y_pred = model.predict(x_test).argmax(axis=-1)
```

```python
# Create figure for displaying predictions
plt.figure(figsize=(8, 8))

# Loop through first 25 test images
for i in range(25):
    
    # Create subplot in 5x5 format
    plt.subplot(5, 5, i+1)
    
    # Remove axis ticks
    plt.xticks([]), plt.yticks([])
    
    # Disable grid
    plt.grid(False)
    
    # Display image
    plt.imshow(x_test[i], cmap='binary')
    
    # Display predicted class name as title
    plt.title(f"Pred: {class_names[y_pred[i]]}")
    
    # Display actual class name as label
    plt.xlabel(f"True: {class_names[y_test[i]]}")

# Show prediction results
plt.show()
```

```python
# Print precision, recall, f1-score, and accuracy for each class
print(classification_report(y_test, y_pred, target_names=class_names))
```

# Overall Working of Project

1. The dataset is loaded using Pandas.
2. Images are reshaped into 28×28 grayscale format.
3. Pixel values are normalized between 0 and 1.
4. A CNN model is created using convolution, pooling, dropout, and dense layers.
5. The model is trained using Fashion MNIST training data.
6. The trained model predicts clothing categories from test images.
7. Accuracy and classification report are used to evaluate performance.
8. Predictions and actual labels are visualized using Matplotlib.
````

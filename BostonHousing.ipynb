````python
# Boston Housing Price Prediction using ANN – Line-by-Line Comments

# Import NumPy library for numerical operations and array handling
import numpy as np

# Import Pandas library for data manipulation and CSV file handling
import pandas as pd

# Import Seaborn library for statistical data visualization
import seaborn as sns

# Import Matplotlib library for plotting graphs and charts
import matplotlib.pyplot as plt

# Import warnings module to manage warning messages
import warnings

# Ignore all warning messages during program execution
warnings.filterwarnings('ignore')
```

```python
# Import function to split dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Import StandardScaler for feature normalization
from sklearn.preprocessing import StandardScaler

# Import Sequential model used to create ANN layer-by-layer
from keras.models import Sequential

# Import Dense layer used in Artificial Neural Network
from keras.layers import Dense
```

```python
# Read Boston Housing dataset CSV file and store it in DataFrame
df = pd.read_csv('BostonHousing.csv')
```

```python
# Display first 5 rows of dataset
df.head()
```

```python
# Display last 5 rows of dataset
df.tail()
```

```python
# Check for missing (null) values in each column
df.isnull().sum()
```

```python
# Display information about dataset such as columns and data types
df.info
```

```python
# Generate statistical summary of dataset
# Includes mean, standard deviation, min, max, etc.
df.describe()
```

```python
# Create figure with size 12x8 inches
plt.figure(figsize=(12, 8))

# Plot correlation heatmap to understand relationships between features
# annot=True displays correlation values on heatmap
# fmt='.2f' shows values up to 2 decimal places
# cmap='coolwarm' sets color style
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
```

```python
# Select input features from dataset
# lstat = lower status population percentage
# ptratio = pupil-teacher ratio
# rm = average number of rooms
x = df.loc[:, ['lstat', 'ptratio', 'rm']]

# Select target/output column
# medv = median house value
y = df.loc[:, "medv"]

# Display shape of input and output data
x.shape, y.shape
```

```python
# Split dataset into training and testing data
# test_size=0.20 means 20% data used for testing
# random_state=10 ensures same random split every time
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.20, random_state=10)
```

```python
# Create object of StandardScaler for normalization
sc = StandardScaler()

# Fit scaler on training data to calculate mean and standard deviation
sc.fit(xtrain)

# Transform training data into normalized form
xtrain = sc.transform(xtrain)

# Transform testing data using same scaler
xtest = sc.transform(xtest)
```

```python
# Create empty Sequential ANN model
model = Sequential()
```

```python
# Create Sequential ANN model
model = Sequential()

# Add first hidden layer with 128 neurons
# input_shape=(3,) because there are 3 input features
# relu activation helps model learn non-linear patterns
model.add(Dense(128, input_shape=(3,), activation='relu'))

# Add second hidden layer with 64 neurons
model.add(Dense(64, activation='relu'))

# Add third hidden layer with 32 neurons
model.add(Dense(32, activation='relu'))

# Add fourth hidden layer with 16 neurons
model.add(Dense(16, activation='relu'))

# Add fifth hidden layer with 8 neurons
model.add(Dense(8, activation='relu'))

# Add output layer with 1 neuron for regression output
# linear activation is used for continuous values
model.add(Dense(1, activation='linear'))

# Compile ANN model
# optimizer='adam' improves learning speed
# loss='mse' calculates Mean Squared Error
# metrics=['mae'] calculates Mean Absolute Error
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Display structure of ANN model
model.summary()
```

```python
# Train ANN model using training dataset
# epochs=100 means model will train 100 times on full dataset
# validation_split=0.05 reserves 5% data for validation
model.fit(xtrain, ytrain, epochs=100, validation_split=0.05)
```

```python
# Evaluate trained model on testing dataset
output = model.evaluate(xtest, ytest)

# Print Mean Squared Error value
print(f"Mean Squared Error: {output[0]}")

# Print Mean Absolute Error value
print(f"Mean Absolute Error: {output[1]}")
```

```python
# Import NumPy library again
import numpy as np

# Create new sample input data for prediction
# Format: [lstat, ptratio, rm]
new_data = np.array([[5, 15, 6]])

# Predict house price using trained ANN model
predictions = model.predict(new_data)

# Display predicted house price
print(predictions)
```

# Overall Working of Project

1. The Boston Housing dataset is loaded using Pandas.
2. Dataset is analyzed using head(), tail(), describe(), and info().
3. Correlation heatmap is used for feature relationship analysis.
4. Important features are selected for prediction.
5. Data is divided into training and testing sets.
6. StandardScaler normalizes input features.
7. An Artificial Neural Network (ANN) model is created using Dense layers.
8. The model is trained using training data.
9. Model performance is evaluated using MSE and MAE.
10. Finally, the model predicts house prices for new input data.
````
